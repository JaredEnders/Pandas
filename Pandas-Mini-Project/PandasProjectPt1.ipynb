```python
# Dependencies
import pandas as pd
```

```python
# load CSV
csv_path = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas
coder_df = pd.read_csv(csv_path)
coder_df.head()
```

```python
# Inspect all columns
coder_df.count()
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
reduced_coder = coder_df.iloc[:,[0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
reduced_coder.head() 
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
replace_list = ["HasChildren", "AttendedBootcamp", "BootcampFinish", "BootcampLoanYesNo", "BootcampRecommend", "BootcampFullJobAfter", "CodeEventBootcamp"]

reduced_coder[replace_list] = reduced_coder[replace_list].replace({0: "No", 1: "Yes"})
reduced_coder.head() 
```

```python
# Calculate total number of respondents in survey
totalrespondents = len(reduced_coder)
totalrespondents
```

```python
# Extract rows corresponding only to people who attended a bootcamp
attended = reduced_coder.loc[reduced_coder["AttendedBootcamp"] == "Yes"]
attended.head()
```

```python
# Calculate average age of attendees
averageage = attended["Age"].mean()
print(averageage)
# Calculate how many people attended a bootcamp
attendance = attended["AttendedBootcamp"].count()
print(attendance)
# Calculate how many attendees hold degrees
degrees = ["bachelor's degree", "master's degree (non-professional)", "professional degree (MBA, MD, JD, etc.)", "associate's degree", "Ph.D."]
collegedegrees = attended["SchoolDegree"].isin(degrees).sum()
print(collegedegrees)
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
attended["Gender"].value_counts()
# Male count
males = attended["Gender"].value_counts()['male']
print(males)
# Female count
females = attended["Gender"].value_counts()['female']
print(females)
# Non-binary count
nb = ["genderqueer", "trans", "agender"]
nonbinary = attended["Gender"].isin(nb).sum()
print(nonbinary) 
# Calculate percentage of respondents who attended a bootcamp
percentattend = round((attendance / totalrespondents) * 100,2)
print(percentattend)
# Calculate percentage of respondents belonging to each gender
genderresp = attended["Gender"].notna().sum()
percmales = round((males / genderresp) * 100,2)
percfemales = round((females / genderresp) * 100,2)
percnb = round((nonbinary / genderresp) * 100,2)
print(percmales)
print(percfemales)
print(percnb)
# Calculate percentage of attendees with a school degree
percdegree = round((collegedegrees / attendance) * 100,2)
print(percdegree) 
# Calculate average post-bootcamp salary
avgsalary = round(attended["BootcampPostSalary"].mean(),0)
print(avgsalary)
```

```python
# Create a new table consolodating above calculations
table_df = {"Total Number of People Surveyed": [totalrespondents], "Total Attendees": [attendance], "Average Age of Attendee": [averageage], "Attendees with Degree": [collegedegrees], "Percent of Attendees with Degree": [percdegree], "Male Attendance": [males], "Percentage of Male Attendees": [percmales], "Female Attendance": [females], "Percentage of Female Attendance": [percfemales], "Non-binary Attendance": [nonbinary], "Non-binary Percentage of Attendees": [percnb], "Average Salary Post-Bootcamp": [avgsalary]}
pd.DataFrame(table_df) 
```

```python
# Improve formatting before outputting spreadsheet
def as_currency(amount):
        return '${:,.2f}'.format(amount)
def as_percent(amount):
    return "{0:.2f}%".format(amount)

table_df = {"Total Number of People Surveyed": [totalrespondents], "Total Attendees": [attendance], "Average Age of Attendee": [averageage], "Attendees with Degree": [collegedegrees], "Percent of Attendees with Degree": as_percent(percdegree), "Male Attendance": [males], "Percentage of Male Attendees": as_percent(percmales), "Female Attendance": [females], "Percentage of Female Attendance": as_percent(percfemales), "Non-binary Attendance": [nonbinary], "Non-binary Percentage of Attendees": as_percent(percnb), "Average Salary Post-Bootcamp": as_currency(avgsalary)}
pd.DataFrame(table_df)
```
