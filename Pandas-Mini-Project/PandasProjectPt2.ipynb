```python
# Dependencies
import pandas as pd
```

```python
# load CSV
csv_path = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
coder_df = pd.read_csv(csv_path, encoding= 'iso-8859-1', low_memory=False)
coder_df.head() 
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
reduced_coder = coder_df.iloc[:,[0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
reduced_coder.head() 
```

```python
# Change reading 0 and 1 to No and Yes, respectively
replace_list = ["HasChildren", "AttendedBootcamp", "BootcampFinish", "BootcampLoanYesNo", "BootcampRecommend", "BootcampFullJobAfter", "CodeEventBootcamp"]

reduced_coder[replace_list] = reduced_coder[replace_list].replace({0: "No", 1: "Yes"})
reduced_coder.head() 
```

```python
# Extract rows for only those who attended a bootcamp
attended = reduced_coder.loc[reduced_coder["AttendedBootcamp"] == "Yes"]
attended
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
difbootcamp = pd.DataFrame(coder_df["BootcampName"].value_counts())
difbootcamp.reset_index(inplace=True)
difbootcamp.columns = ["BootcampName", "Number of Attendees"]
difbootcamp.head() 
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
recommended_df = attended.replace({"Yes": 1, "No": 0})
recommended_df = pd.DataFrame(coder_df.groupby("BootcampName")["BootcampRecommend"].sum())
recommended_df.reset_index(inplace=True)
recommended_df.columns = ["BootcampName", "Number of Recommenders"]
recommended_df
```

```python
# Merge the two created data frames on the name of the bootcamp
merge_df = pd.merge(difbootcamp, recommended_df, on= "BootcampName")
merge_df
```

```python
# Calculate percentage of each bootcamp's students who are recommenders
merge_df["Percentage of People Recommended"] = merge_df["Number of Recommenders"] / merge_df["Number of Attendees"] *100
merge_df["Percentage of People Recommended"] = merge_df["Percentage of People Recommended"].map("{0:,.2f}%".format)
merge_df
```

```python
# Export to excel and remove index
PandasProjectDataPt2 = pd.DataFrame(merge_df)
PandasProjectDataPt2.to_excel("/Users/jared/Pandas/Pandas-Mini-Project/PandasDataPt2.xlsx", index=False)
```
